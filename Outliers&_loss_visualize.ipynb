```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Mean Squared Error (MSE) loss
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# Function to calculate Smooth L1 Loss
def smooth_l1_loss(errors):
    return np.where(np.abs(errors) < 1, 0.5 * errors**2, np.abs(errors) - 0.5)

# Generate some example data
np.random.seed(42)
y_true = np.random.rand(100)  # Ground truth values
y_pred = y_true + np.random.normal(0, 0.2, 100)  # Simulated predicted values with noise

# Calculate errors
errors = y_true - y_pred

# Calculate losses
mse_loss_value = mse_loss(y_true, y_pred)
smooth_l1_loss_values = smooth_l1_loss(errors)

# Plotting the losses
plt.figure(figsize=(12, 6))

# Plot Mean Squared Error (MSE) loss
plt.subplot(1, 2, 1)
plt.scatter(y_true, y_pred, label=f'MSE Loss: {mse_loss_value:.4f}')
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.title('Mean Squared Error (MSE) Loss')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.legend()

# Plot Smooth L1 Loss
plt.subplot(1, 2, 2)
plt.plot(errors, smooth_l1_loss_values, label='Smooth L1 Loss')
plt.title('Smooth L1 Loss Function')
plt.xlabel('Errors')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
